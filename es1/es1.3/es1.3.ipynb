In questo esercizio si stima $\pi$ con l'esperimento di Buffon che sfutta la formula:  
$$\pi = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$  
dove L è la lunghezza di un ipotetico ago lanciato randomicamente in un campo su cui vi è un fascio di linee parallele a distanza d>L.  
 

Poichè devo stimare $\pi$ ho evitato di usare il valore il suo valore all'interno del codice. Non ho quindi rappresentato l'ago tramite cordinate polari ma ho estratto la posizione di una delle due estremità e poi ho estratto un vettore di modulo d.  
Per fare ciò ho creato in: "random.cpp" la funzione: "vers_2D()" che mi estrae un punto all'interno di un quadrato unitario finchè questo non si trova all'interno di una circonferenza unitaria, quando ciò accade questo mi viene restituito. 

```python
import matplotlib.pyplot as plt
import numpy as np

M=10000000             # numero di tiri
N=100                 # Numero di blocchi
pi_teo=3.14159265358979323846
L=int(M/N)
xx = np.arange(N)      # [0,1,2,...,N-1]
# print (type(f))

with open("out_err", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    err=[]
    for i in righe:
            if(i!=""):
                try:
                    # print(i.split("\t"))
                    x,y=i.split("\t")
                    # print(i)
                    err.append(float(x))
                except ValueError:
                    print(f"Il valore '{i}' non può essere convertito in float.")
        
    # print(err)
with open("out_pi", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    pi=[]
    for i in righe:
            if(i!=""):
                try:
                    # print(i.split("\t"))
                    x,y=i.split("\t")
                    # print(i)
                    pi.append(float(x))
                except ValueError:
                    print(f"Il valore '{i}' non può essere convertito in due float.")

xx=xx+1
bl=xx
# print(bl)
plt.plot(bl,pi)
plt.xlabel('#blocchi')
plt.ylabel(r'$\pi$')
plt.grid(True)
plt.show()
plt.subplots()
plt.plot(bl,err)
plt.xlabel('#blocchi')
plt.ylabel(r'$\sigma_{\pi}$')
plt.grid(True)
plt.show()
xx*=L # Number of throws = block * (Number of throws in each block)
figure,ass=plt.subplots()
plt.errorbar(xx,np.array(pi)-pi_teo,yerr=err)
plt.xlabel('#tiri')
plt.ylabel(r'$\pi-\pi_{teo}$')
plt.grid(True)
plt.show()
```

Con l'aumentare del numero di blocchi la media a blocchi converge al un valore compatibile con il valore esatto di $\pi$  
($\pi_{teo}=3.14159265358979323846...$).
