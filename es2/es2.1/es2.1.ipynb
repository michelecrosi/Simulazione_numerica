L'obbiettivo dell'esercizio è quello di stimare l'integrale:  
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$$  
il cui risultato analitico è $I=1$.  
In generale per stimare $I'=\int_a^b f(x) dx$ si usa:  
$$I'=\frac{1}{N}\sum^N_{i=0} \frac{f}{p}(x_i)$$ 
con $x_i$ che segue la distribuzione p(x).

Ho usato $p(x)=1+\frac{\pi}{4}-\frac{\pi}{2}x$ con $x\in [0,1)$.  
Questa è la distribuzione che si ottiene normalizzando in $x\in[0;1]$ lo sviluppo di Taylor di: $\frac{\pi}{2}\cos(\pi x/2)$  centrato in $\frac{1}{2}$.  
Questa distribuzione non è mai nulla nel dominio, il che è necessario in quanto p(x) appare al denominatore della formula per stimare I.   Essendo una distribuzione è sempre positiva nel dominio

Usando $x=F_x^{-1}(y)$ e facendo i calcoli a mano ho trovato la seguente formula per estrarre con una distribuzione lineare la variabile $x\in[a;b]$:  
$p(x)=mx+q \Rightarrow x=\frac{-q+\sqrt{(q+a m)^2+2 m y}}{m}$
