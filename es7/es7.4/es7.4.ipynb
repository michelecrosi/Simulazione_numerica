## salvataggio dati

```python
# immagazzinamento dati

import matplotlib.pyplot as plt
import numpy as np


titol=["solido","liquido","gas"]
propiet=["potential_energy","pressure","gofr"]
ERR=[]
X=[]
Y=[]
Y_AV=[]
# il primo indice è la fase, il secondo è la proprietà misurata, il terzo è il numero del blocco (nel caso di gofr invece è la distanza)	


nblk=-1 #posso mettere un valore qualsiasi iniziale

A=0
for i in titol:
	ERR.append([])
	X.append([])
	Y.append([])
	Y_AV.append([])
	B=0
	for j in propiet:
		with open(f"./../es7.3/{i}/{j}.dat", 'r') as f:
			righe=f.read().split("\n")
			ERR[A].append([])
			X[A].append([])
			Y[A].append([])
			Y_AV[A].append([])
			for l in righe:
				if(l!=""):
					try:
						if(j=="gofr"):
							# print("\n\n")
							a,b,c,d,e=l.split() #split() separa se ho cose tra 1 o più spazi vuoti
							# print(a)
							# print(nblk)
							# print(int(a)==nblk)
							if(int(a)==nblk): # salvo solo gli ultimi risultati
								# print("ffjakljfòaslkfa")
								X[A][B].append(float(b))
								Y[A][B].append(float(c))
								Y_AV[A][B].append(float(d))
								ERR[A][B].append(float(e))
						else:
							a,b,c,d=l.split()
							X[A][B].append(float(a))
							Y[A][B].append(float(b))
							Y_AV[A][B].append(float(c))
							ERR[A][B].append(float(d))
							# print(b)	
					except ValueError:
						print(f"Il valore '{l}' non può essere convertito in floats.")
		B+=1
		nblk=len(X[0][0])	
		# print(nblk)

	A+=1



# ora i dati con la dinamica molecolare:

titol=["solido","liquido","gas"]
propiet=["potential_energy","pressure","gofr"]
ERR_DM=[]
X_DM=[]
Y_DM=[]
Y_AV_DM=[]
# il primo indice è la fase, il secondo è la proprietà misurata, il terzo è il numero del blocco (nel caso di gofr invece è la distanza)	


nblk=-1 #posso mettere un valore qualsiasi iniziale

A=0
for i in titol:
	ERR_DM.append([])
	X_DM.append([])
	Y_DM.append([])
	Y_AV_DM.append([])
	B=0
	for j in propiet:
		with open(f"./{i}/{j}.dat", 'r') as f:
			righe=f.read().split("\n")
			ERR_DM[A].append([])
			X_DM[A].append([])
			Y_DM[A].append([])
			Y_AV_DM[A].append([])
			for l in righe:
				if(l!=""):
					try:
						if(j=="gofr"):
							# print("\n\n")
							a,b,c,d,e=l.split() #split() separa se ho cose tra 1 o più spazi vuoti
							# print(a)
							# print(nblk)
							# print(int(a)==nblk)
							if(int(a)==nblk): # salvo solo gli ultimi risultati
								# print("ffjakljfòaslkfa")
								X_DM[A][B].append(float(b))
								Y_DM[A][B].append(float(c))
								Y_AV_DM[A][B].append(float(d))
								ERR_DM[A][B].append(float(e))
						else:
							a,b,c,d=l.split()
							X_DM[A][B].append(float(a))
							Y_DM[A][B].append(float(b))
							Y_AV_DM[A][B].append(float(c))
							ERR_DM[A][B].append(float(d))
							# print(b)	
					except ValueError:
						print(f"Il valore '{l}' non può essere convertito in floats.")
		B+=1
		nblk=len(X_DM[0][0])	
		# print(nblk)

	A+=1
```

## grafici

### microcanonico

```python
# test

import matplotlib.pyplot as plt
import numpy as np

titol=["solido","liquido","gas"]

TERR_DM=[]
TX_DM=[]
TY_DM=[]
TY_AV_DM=[]

A=0
for i in titol:
	TERR_DM.append([])
	TX_DM.append([])
	TY_DM.append([])
	TY_AV_DM.append([])
	with open(f"./{i}/temperature.dat", 'r') as f:
		righe=f.read().split("\n")
		for l in righe:
			if(l!=""):
				try:
					a,b,c,d=l.split()
					TX_DM[A].append(float(a))
					TY_DM[A].append(float(b))
					TY_AV_DM[A].append(float(c))
					TERR_DM[A].append(float(d))
						# print(b)	
				except ValueError:
					print(f"Il valore '{l}' non può essere convertito in floats.")
	A+=1

A=0
for i in titol:
	plt.errorbar(TX_DM[A], TY_AV_DM[A], np.array(TERR_DM[A]))	
	plt.title(f"{i}:")
	plt.xlabel("#blocchi")
	plt.ylabel("T")
	plt.grid(True)
	plt.show()
	A+=1
```

### microcanonico vs canonico

Compariamo la g(r) dell'insieme canonico con quella dell'insieme microcanonico a parità di densità, temperatura e numero di particelle.

1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

```python
# da eseguire dopo il salvataggio dati

A=0
for i in titol:
	plt.errorbar(X[A][2], Y_AV[A][2], yerr=ERR[A][2], fmt='o', color='blue', label='Canonico', ls='-', lw=1, markersize=4)
	plt.errorbar(X_DM[A][2], Y_AV_DM[A][2], yerr=ERR_DM[A][2], fmt='o', color='red', label='Microcanonico', ls='-', lw=1, markersize=4)
	plt.legend()
	
	plt.title(f"{i}:")
	plt.xlabel("r")
	plt.ylabel("<g(r)>")
	plt.grid(True)
	plt.show()
	A+=1
```

Ora grafico la differenza delle due $<g(r)>$ associandogli come errore la somma in quadratura degli errori delle due funzioni.

```python
# da eseguire dopo il salvataggio dati

for A in range(len(titol)):
	plt.errorbar(X[A][2], np.array(Y_AV[A][2])-np.array(Y_AV_DM[A][2]), yerr=np.sqrt(np.power(np.array(ERR[A][2]),2)+np.power(np.array(ERR_DM[A][2]),2)), fmt='o', color='blue', ls='-', lw=1, markersize=4)
	plt.title(f"{titol[A]}:")
	plt.xlabel("r")
	plt.ylabel(r"$<g(r)>_{MC}-<g(r)>_{DM}$") #montecarlo meno dinamica molecolare
	plt.grid(True)
	plt.show()
```

La g(r) campionata nell'insieme canonico risulta molto simile a quella campionata nelle stesse condizioni nell'insieme microcanonico.  
Tuttavia la differenza tra le due funzione non sembra compatibile con la funzione identicamente nulla, questo può essere dovuto al fatto che l'uguaglianza tra i due insiemi vi è solo nel limite termodinamico.
Un altro motivo può essere che le condizioni non sono esattamente le stesse in quanto la temperatura nell'insieme microcanonico non è fissata e quindi la temperatura d'equilibrio non può essere scelta con precisione.
