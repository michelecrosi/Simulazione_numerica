```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit # per il fit

# funzione di fitting
def func(x, k):
    return k * np.sqrt(x)

with open("out_RW_discr_err", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    err=[]
    xx=[]
    for i in righe:
            if(i!=""):
                try:
                    # print(i.split("\t"))
                    x,y=i.split("\t")
                    # print(i)
                    err.append(float(x))
                    xx.append(int(y))
                except ValueError:
                    print(f"Il valore '{i}' non puÃ² essere convertito in float.")
        
print(err)
with open("out_RW_discr", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    RW=[]
    for i in righe:
            if(i!=""):
                try:
                    x,y=i.split("\t")
                    # print(i)
                    RW.append(float(x))
                except ValueError:
                    print(f"Il valore '{i}' non puÃ² essere convertito in due float.")

popt, pcov = curve_fit(func, xx, RW, sigma=err, absolute_sigma=True)

# Recupero il valore di k ottimale
k_opt = popt[0]

# Genero la curva fittata
x_fit = np.linspace(1, 10, 100)
y_fit = func(x_fit, k_opt)

# Creo il grafico
plt.figure(figsize=(8, 6))
plt.errorbar(xx, RW, yerr=err, fmt='o', label='Dati', color='blue', ecolor='black', elinewidth=3, capsize=0)
plt.plot(x_fit, y_fit, label=f'Fit: $y = {k_opt:.2f} \sqrt{{x}}$', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Fit dei dati con $f(x) = k \sqrt{x}$')
plt.show()


with open("out_RW_cont_err", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    err=[]
    xx=[]
    for i in righe:
            if(i!=""):
                try:
                    # print(i.split("\t"))
                    x,y=i.split("\t")
                    # print(i)
                    err.append(float(x))
                    xx.append(int(y))
                except ValueError:
                    print(f"Il valore '{i}' non puÃ² essere convertito in float.")
        
    # print(err)
with open("out_RW_cont", 'r') as f:# con with il file si chiude da solo appena finito 
    righe=f.read().split("\n")
    # print(righe)
    RW=[]
    for i in righe:
            if(i!=""):
                try:
                    # print(i.split("\t"))
                    x,y=i.split("\t")
                    # print(i)
                    RW.append(float(x))
                except ValueError:
                    print(f"Il valore '{i}' non puÃ² essere convertito in due float.")

popt, pcov = curve_fit(func, xx, RW, sigma=err, absolute_sigma=True)

# Recupero il valore di k ottimale
k_opt = popt[0]

# Genero la curva fittata
x_fit = np.linspace(1, 10, 100)
y_fit = func(x_fit, k_opt)

# Creo il grafico
plt.figure(figsize=(8, 6))
plt.errorbar(xx, RW, yerr=err, fmt='o', label='Dati', color='blue', ecolor='black', elinewidth=3, capsize=0)
plt.plot(x_fit, y_fit, label=f'Fit: $y = {k_opt:.2f} \sqrt{{x}}$', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Fit dei dati con $f(x) = k \sqrt{x}$')
plt.show()
```
