Assumiamo che il prezzo di un bene: S(t) al variare del tempo segua un moto Browniano Geometrico con volatilità: $\sigma=0.25$, tasso di interesse: $r=0.1$ e prezzo iniziale: $S(t=0)=100$.  
L'obbiettivo di questo esercizio è quello di calcolare il prezzo a cui bisogna vendere il diritto di call ($C[S(t=0),t=0]$) e quello di call ($P[S(t=0),t=0]$) se il questo diritto garantisce il prezzo $K=100$ al tempo $T=1$.  
Questi prezzi sono stimati basandosi sul fatto che il detentore del diritto (holder) agirà nel suo interesse e quindi si avvarrà del diritto solo se questo gli conviene.  
Valgono quindi:  
$C[S(t=0),t=0]=\exp(-rT)\max\{0, S(t=T)-K\}$  
$P[S(t=0),t=0]=\exp(-rT)\max\{0, K-S(t=T)\}$

Per simulare il moto Browniano Geometrico discretizzo il tempo in n intervalli e sfrutto che un moto di questo tipo può essere estratto a partire dalla formula:  
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
con $Z_i \sim \mathcal{N}(0,1)$  
(estratta con distribuzione gaussiana con media 0 e varianza 1)  
con $i\in \{0,...,n-1\}$ e $t_{n}=T$
